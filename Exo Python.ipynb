```python
def remove_adjacent(nums):
    # +++your code here+++
    nums.sort()
    result=[]
    if nums!=[]:
        result.append(nums[0])
        for i in range(1,len(nums)):
            if nums[i-1]!= nums[i]:
                result.append(nums[i])
    return result
```

```python
def linear_merge(list1, list2):
    # +++your code here+++
    result=[]
    result=list1+list2
    result.sort()
    return result
```
